```python
def plot_hist(_data, _id, _w, _h):   
    fig, ax = plt.subplots(_w, _h, figsize=(20, 4*_w))
    for i in range(_w):
        for j in range(_h):
            idx = i*_w+j
            if _w < 2:
                sns.distplot(_data[:,idx], ax=ax[idx])
                ax[idx].set_title(_id[idx])
            else: 
                sns.distplot(_data[:,idx], ax=ax[i][j])
                ax[i][j].set_title(_id[idx])
    plt.show()
```
