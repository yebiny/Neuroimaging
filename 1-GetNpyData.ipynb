```python
BASE_PATH = './'

print('Reading data...')
loading_data = pd.read_csv(f'{BASE_PATH}/data/loading.csv')
trainscores_data = pd.read_csv(f'{BASE_PATH}/data/train_scores.csv')
fnc_data = pd.read_csv(f'{BASE_PATH}/data/fnc.csv')
```

```python
def checking_missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    missing_data  = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    display(missing_data.head())
    
checking_missing_data(fnc_data)
checking_missing_data(loading_data)
checking_missing_data(trainscores_data)
```
